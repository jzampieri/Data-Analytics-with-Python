```python
# 1.Quem venceu a temporada de 2008?

winner2008 = tabela[['season', 'place', 'team', 'acronym']] #puxando dados da tabela
winner_2008 = winner2008.loc[(winner2008['season'] == 2008) & (winner2008['place'] == 1) & (winner2008['team']) & (winner2008['acronym'])]

print(f"{winner_2008}")
```

```python
# 2.Qual time fez o maior número de pontos de todos os tempos? Especifique a temporada e o número de pontos.

pointsall = tabela.sort_values(by='points', ascending=False) #puxando dados da tabela
mostpoints = pointsall.iloc[0]

print(mostpoints)
```

```python
# 3.Qual time teve o melhor aproveitamento de gols? Especifique a temporada e o número.

saldoall = tabela.sort_values(by='goals_diff', ascending=False) #puxando dados da tabela
mostsaldo = saldoall.iloc[0]

print(mostsaldo)
```

```python
# 4.Qual time sofreu mais gols em todos os tempos? Especifique a temporada e o número de gols.

sofreuall = tabela.sort_values(by='goals_against', ascending=False) #puxando dados da tabela
mostsofrido = sofreuall.iloc[0]

print(mostsofrido)
```

```python
# 5.Qual time perdeu mais partida? Especifique a temporada.

lostall = tabela.sort_values(by='loss', ascending=False) #puxando dados da tabela
mostlost = lostall.iloc[0]

print(mostlost)
```

```python
# 6.Qual time teve o pior aproveitamento de gols? Especifique a temporada e o número.

saldopiorall = tabela.sort_values(by='goals_diff', ascending=True) #puxando dados da tabela
mostpiorsaldo = saldopiorall.iloc[0]

print(mostpiorsaldo)
```

```python
# 7.Quem venceu a temporada de 2023?

winner2023 = tabela[['season', 'place', 'team', 'acronym']] #puxando dados da tabela
winner_2023 = winner2023.loc[(winner2023['season'] == 2023) & (winner2023['place'] == 1) & (winner2023['team']) & (winner2023['acronym'])]

print(f"{winner_2023}")
```

```python
# 8.Quem venceu a temporada de 2018?

winner2018 = tabela[['season', 'place', 'team', 'acronym']] #puxando dados da tabela
winner_2018 = winner2018.loc[(winner2018['season'] == 2018) & (winner2018['place'] == 1) & (winner2018['team']) & (winner2018['acronym'])]

print(f"{winner_2018}")
```

```python
# 9.Quem teve mais derrotas em 2020?
data_2020 = tabela[tabela['season'] == 2020]
brvalue = data_2020.sort_values(by='loss', ascending=False)
timemaisderrotas = brvalue.iloc[0]

print(timemaisderrotas)
```

```python
# 10.Quem teve mais vitórias em 2009?

data_2009 = tabela[tabela['season'] == 2009]
brvalue = data_2009.sort_values(by='won', ascending=False)
timemaisvitorias = brvalue.iloc[0]

print(timemaisvitorias)
```
